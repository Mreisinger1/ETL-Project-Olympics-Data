```python
import pandas as pd
from sqlalchemy import create_engine, inspect
```

```python
population_file = "Resources/population .csv"
population_df = pd.read_csv(population_file)
population_df.head()
```

```python
df = population_df.drop(population_df.columns[[1,2,3]], axis=1).drop(population_df.index[[0,1,2]])
df.head()
```

```python
df.columns = df.iloc[0]
df.head()
```

```python
df.columns.tolist()
```

```python
population_m_df = df.melt(id_vars = 'Country Name', value_vars =None, var_name = 'Year')
population_m_df.head()
```

```python
new_pop_df= population_m_df.drop(population_m_df.index[[0]])
new_pop_df.head()
```

```python
new_pop_df.rename(columns={'Country Name' : 'country'}, inplace=True)
new_pop_df.rename(columns={'Year' : 'year'}, inplace=True)
new_pop_df.rename(columns={'value' : 'population'}, inplace=True)
new_pop_df.head()
```

```python
new_pop_df['year'] = new_pop_df['year'].astype(int)
new_pop_df.head()
```

```python
protocol = 'postgresql'
username = 'postgres'
password = 'bootcamp'
host = 'localhost'
port = 5432
database_name = 'olympic_db'
rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
```

```python
inspector = inspect(engine)
inspector.get_table_names()
```

```python
new_pop_df.to_sql(name='country_data', con=engine, if_exists='append', index=False)
```
